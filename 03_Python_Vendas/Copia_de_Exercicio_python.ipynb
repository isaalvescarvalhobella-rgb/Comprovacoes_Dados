Agrupamentos

```python
#media de quantidade vendida por categoria
print(df.groupby('categoria')['quantidade'].mean())
```

```python
#Soma de quantidade vendida por categoria
print(df.groupby('categoria')['quantidade'].sum())
```

```python
#multiplas estatisticas por categoria
print(df.groupby('categoria')['quantidade'].agg(['mean', 'sum', 'count', 'max', 'min']))
```

```python
# 1. Corrigir e converter a coluna preco_unitario
# Assume-se que a vírgula é o separador decimal e o ponto é o separador de milhares.
df['preco_unitario'] = df['preco_unitario'].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
```

.astype(str) `#Garante que é string`
.str.replace('.', '', regex=False) `#remove ponto dos milhares`

.str.replace(',', '.', regex=False) `#troca virgula decimal por ponto`

```python
#converter para float
df['preco_unitario'] = pd.to_numeric(df['preco_unitario'], errors='coerce')
```

```python
#Converter quantidade, se necesario
df['quantidade'] = pd.to_numeric(df['quantidade'], errors='coerce')
```

```python
#multiplicar
df['total_venda'] = df['quantidade'] * df['preco_unitario']
```

```python
#verificar tipo final
print(df['total_venda'].dtype)
```

Criar coluna nova VIP oi Nao VIP

```python
#ciar uma nova coluna Status Do Cliente
df['STATUS']= np.where(df['total_venda'] > df['total_venda'].mean(), 'VIP', 'Nao VIP')
```

```python
#Verificar a distribuição de clientes VIP e Não VIP
print(df['STATUS'].value_counts())
```

Trabalhando com datas

```python
#Garantir que informacoes de data estao no formato datetime
df['data_venda'] = pd.to_datetime(df['data_venda'])
```

```python
#extrair as informacoes de data
df['ano'] = df['data_venda'].dt.year
df['mes'] = df['data_venda'].dt.month
df['dia'] = df['data_venda'].dt.day
df['dia da semana'] = df['data_venda'].dt.dayofweek
```
